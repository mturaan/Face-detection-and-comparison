```python

import cv2
import face_recognition
from google.colab.patches import cv2_imshow #importing for only google colab

image = cv2.imread("drive/MyDrive/Colab Notebooks/face_rec/2.jpg")
path="drive/MyDrive/Colab Notebooks/face_rec/1.jpg"
test = face_recognition.load_image_file(path)
testEncoding=face_recognition.face_encodings(test)[0]  #For one face index is=0


testPath = "drive/MyDrive/Colab Notebooks/face_rec/2.jpg"
testImage = face_recognition.load_image_file(testPath)
faceLoc= face_recognition.face_locations(testImage)
faceEnc=face_recognition.face_encodings(testImage,faceLoc)

#print(faceLoc) # to determine the location of face. 
topLeftX= 270    #index 3
topLeftY= 116    #index 0
bottomRightX=425 #index 1
bottomRightY=270 #index 2
matchedFaces=face_recognition.compare_faces(testEncoding,faceEnc)
if True in matchedFaces:
  cv2.rectangle(image,(topLeftX,topLeftY),(bottomRightX,bottomRightY),(0,255,0),(2))
  cv2.putText(image,"person",(topLeftX,topLeftY),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),1)
  cv2_imshow(image)
else:
  cv2.rectangle(image, (topLeftX,topLeftY),(bottomRightX,bottomRightY),(0,0,191),(3)) #BGR
  cv2.putText(image,"unknown",(topLeftX,topLeftY),cv2.FONT_HERSHEY_SIMPLEX,2,(0,0,191),2) #BGR
  cv2_imshow(image)
```
